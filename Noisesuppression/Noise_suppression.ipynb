```python
# 1. Install required packages
!pip install noisereduce librosa soundfile git+https://github.com/openai/whisper.git

import librosa
import soundfile as sf
import noisereduce as nr
import whisper
import IPython.display as ipd

# 2. Load the raw audio file
# Replace 'path_to_your_audio.m4a' with the uploaded file path
audio_path = '/content/AUD-20260402-WA0018.m4a'
y, sr = librosa.load(audio_path, sr=None)

print("Original Audio Quality Processing...")

# 3. Apply Non-Stationary Noise Reduction
# Set stationary=False to dynamically adapt to shifting background chatter
y_denoised = nr.reduce_noise(y=y, sr=sr, stationary=False, prop_decrease=0.85)

# 4. Save the clean output file
output_path = 'cleaned_discussion_audio.wav'
sf.write(output_path, y_denoised, sr)
print(f"Noise reduction complete! Saved to {output_path}")

# 5. Optional: Listen to the audio directly in Colab
print("Listen to Cleaned Audio:")
ipd.display(ipd.Audio(output_path))

```
